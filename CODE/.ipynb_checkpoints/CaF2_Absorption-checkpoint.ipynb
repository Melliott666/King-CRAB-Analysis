```python
# Nexus did not have the absorption length information for Calcium Fluoride and all our optical pathing needs this
# Görling https://www.sciencedirect.com/science/article/pii/S0030401805000490?via%3Dihub has the bulk behavior
# For two wavelengths
alpha_193_bulk = 2.89e-3
alpha_157_bulk = 17.5e-3

# Zukic https://opg.optica.org/ao/fulltext.cfm?uri=ao-29-28-4284 has the thin film behavior as a broad spectra
ZUKIC_JSON = Path("DATA/caf2_zukic_k.json")
# So I will take that behavior, fit it to mach the bulk behavior and export this to Nexus

# ----------------------------
# Load Zukic digitized data: use pt["value"] = [lambda_nm, k]
# ----------------------------
with open(ZUKIC_JSON, "r") as f:
    j = json.load(f)

pts = j["datasetColl"][0]["data"]
lambda_nm = np.array([p["value"][0] for p in pts], dtype=float)
k = np.array([p["value"][1] for p in pts], dtype=float)

# sort
idx = np.argsort(lambda_nm)
lambda_nm = lambda_nm[idx]
k = k[idx]

if np.any(k <= 0):
    raise ValueError("k has non-positive entries; check digitization/log-axis settings.")

# ----------------------------
# Plot 1: Input k(λ) with LINEAR interpolation between points
# ----------------------------
lam_dense = np.linspace(lambda_nm.min(), lambda_nm.max(), 2000)
k_dense_linear = interp_linear(lam_dense, lambda_nm, k)

plt.figure()
plt.plot(lambda_nm, k, marker="o", linestyle="None")
plt.plot(lam_dense, k_dense_linear, linestyle="-")
plt.yscale("log")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Extinction coefficient k")
plt.title("Zukic CaF2 k(λ): points + linear interpolation")
plt.grid(True, which="both", alpha=0.3)
```

```python



energy_eV = np.array([
    0.900, 1.000, 1.296, 1.683, 2.075,
    2.585, 3.088, 3.709, 4.385, 4.972,
    5.608, 6.066, 6.426, 6.806, 7.135,
    7.401, 7.637, 7.880, 8.217
], dtype=float)

ABSLENGTH_LONG_MM = 1.0e5  # 100 m for λ > 230 nm (transparent region)
ALLOW_EXTRAPOLATION = True

# ----------------------------
# Helpers
# ----------------------------
def energy_to_lambda_nm(E):
    return 1239.841984 / E

def alpha_from_k(lambda_nm, k):
    # α(cm^-1) = 4π k / λ(cm), λ(cm) = λ_nm * 1e-7
    return 4.0 * np.pi * k / (lambda_nm * 1e-7)

def interp_logy(xq, x, y):
    # log(y) linear in x interpolation (no extrapolation)
    y = np.asarray(y, dtype=float)
    if np.any(y <= 0):
        raise ValueError("interp_logy requires y > 0")
    return np.exp(np.interp(xq, x, np.log(y)))

def extrap_logy(xq, x, y):
    # log(y) linear in x interpolation with endpoint extrapolation
    x = np.asarray(x, dtype=float)
    y = np.asarray(y, dtype=float)
    if np.any(y <= 0):
        raise ValueError("extrap_logy requires y > 0")

    logy = np.log(y)
    xq = np.atleast_1d(xq).astype(float)
    out = np.empty_like(xq)

    # endpoint slopes
    m_lo = (logy[1] - logy[0]) / (x[1] - x[0])
    b_lo = logy[0] - m_lo * x[0]
    m_hi = (logy[-1] - logy[-2]) / (x[-1] - x[-2])
    b_hi = logy[-1] - m_hi * x[-1]

    for i, xv in enumerate(xq):
        if xv < x[0]:
            out[i] = np.exp(m_lo * xv + b_lo)
        elif xv > x[-1]:
            out[i] = np.exp(m_hi * xv + b_hi)
        else:
            out[i] = np.exp(np.interp(xv, x, logy))
    return out

def format_cpp_vector(name, values_mm):
    parts = [f"{v:.6g}*mm" for v in values_mm]
    return "std::vector<G4double> {} = {{\n  {}\n}};\n".format(name, ",\n  ".join(parts))

# ----------------------------
# Load Zukic digitized data (use value[0], value[1])
# ----------------------------
with open(ZUKIC_JSON, "r") as f:
    j = json.load(f)

data = j["datasetColl"][0]["data"]

lambda_nm = np.array([pt["value"][0] for pt in data], dtype=float)
k         = np.array([pt["value"][1] for pt in data], dtype=float)

# Sort by wavelength
order = np.argsort(lambda_nm)
lambda_nm = lambda_nm[order]
k = k[order]

if np.any(lambda_nm <= 0) or np.any(k <= 0):
    raise ValueError("Digitized lambda_nm and k must be > 0. Check your export.")

# ----------------------------
# Shape α from Zukic
# ----------------------------
alpha_shape = alpha_from_k(lambda_nm, k)  # cm^-1

# Evaluate α_shape at 193 and 157 nm
if ALLOW_EXTRAPOLATION:
    alpha_shape_193 = extrap_logy(193.0, lambda_nm, alpha_shape)[0]
    alpha_shape_157 = extrap_logy(157.0, lambda_nm, alpha_shape)[0]
else:
    alpha_shape_193 = interp_logy(193.0, lambda_nm, alpha_shape)
    alpha_shape_157 = interp_logy(157.0, lambda_nm, alpha_shape)

# Ratios to force bulk truth at anchors
r193 = alpha_193_bulk / alpha_shape_193
r157 = alpha_157_bulk / alpha_shape_157

log_r193 = np.log(r193)
log_r157 = np.log(r157)

def r_of_lambda(lam_nm):
    # log-linear between 193 and 157 nm, extrapolated outside
    lam_nm = np.asarray(lam_nm, dtype=float)
    t = (lam_nm - 193.0) / (157.0 - 193.0)  # 0 at 193, 1 at 157
    return np.exp(log_r193 + t * (log_r157 - log_r193))

alpha_corr_curve = alpha_shape * r_of_lambda(lambda_nm)

# Sanity check: corrected curve hits targets at anchors
if ALLOW_EXTRAPOLATION:
    a193 = extrap_logy(193.0, lambda_nm, alpha_corr_curve)[0]
    a157 = extrap_logy(157.0, lambda_nm, alpha_corr_curve)[0]
else:
    a193 = interp_logy(193.0, lambda_nm, alpha_corr_curve)
    a157 = interp_logy(157.0, lambda_nm, alpha_corr_curve)

print("Anchor check (should match):")
print(f"  alpha_corr(193) = {a193:.3e} cm^-1   target {alpha_193_bulk:.3e}")
print(f"  alpha_corr(157) = {a157:.3e} cm^-1   target {alpha_157_bulk:.3e}")
print(f"  L(193) = {10.0/alpha_193_bulk:.1f} mm,  L(157) = {10.0/alpha_157_bulk:.1f} mm")
print()

# ----------------------------
# Evaluate on your energy grid
# ----------------------------
lambda_grid = energy_to_lambda_nm(energy_eV)
alpha_grid = np.empty_like(lambda_grid)

lam_min, lam_max = float(lambda_nm[0]), float(lambda_nm[-1])

for i, lam in enumerate(lambda_grid):
    if lam > lam_max:
        # long wavelength region not covered by Zukic: treat as essentially transparent
        alpha_grid[i] = 1.0 / (ABSLENGTH_LONG_MM / 10.0)  # mm->cm then invert
    else:
        if ALLOW_EXTRAPOLATION:
            alpha_grid[i] = extrap_logy(lam, lambda_nm, alpha_corr_curve)[0]
        else:
            if lam < lam_min:
                raise ValueError(f"Need extrapolation at λ={lam:.2f} nm but ALLOW_EXTRAPOLATION=False")
            alpha_grid[i] = interp_logy(lam, lambda_nm, alpha_corr_curve)

L_mm = 10.0 / alpha_grid  # since L_cm=1/alpha, L_mm=10*L_cm

print("Energy (eV) -> lambda (nm) -> ABSLENGTH (mm):")
for E, lam, L in zip(energy_eV, lambda_grid, L_mm):
    print(f"  {E:6.3f} eV  {lam:8.2f} nm  {L:12.3g} mm")
print()

print("C++ vector (internal points only):\n")
print(format_cpp_vector("caf2_abs_length", L_mm))
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# ----------------------------
# Settings
# ----------------------------
ZUKIC_JSON = Path("/Users/physics/Downloads/caf2_zukic_k.json")

alpha_193_bulk = 2.89e-3   # cm^-1 (Görling 2005)
alpha_157_bulk = 17.5e-3   # cm^-1 (Görling 2005)

energy_eV = np.array([
    0.900, 1.000, 1.296, 1.683, 2.075,
    2.585, 3.088, 3.709, 4.385, 4.972,
    5.608, 6.066, 6.426, 6.806, 7.135,
    7.401, 7.637, 7.880, 8.217
], dtype=float)

ABSLENGTH_LONG_MM = 1.0e4      # 100 m for wavelengths beyond Zukic max (~230 nm)
ALLOW_EXTRAPOLATION = True     # allow small extrapolation beyond digitized range

# ----------------------------
# Helpers
# ----------------------------
def energy_to_lambda_nm(E):
    return 1239.841984 / E

def alpha_from_k(lambda_nm, k):
    # α(cm^-1) = 4π k / λ(cm), λ(cm)=λ_nm*1e-7
    return 4.0 * np.pi * k / (lambda_nm * 1e-7)

def interp_linear(xq, x, y):
    return np.interp(xq, x, y)

def interp_logy(xq, x, y):
    y = np.asarray(y, float)
    if np.any(y <= 0):
        raise ValueError("interp_logy requires y>0")
    return np.exp(np.interp(xq, x, np.log(y)))

def extrap_logy(xq, x, y):
    x = np.asarray(x, float)
    y = np.asarray(y, float)
    if np.any(y <= 0):
        raise ValueError("extrap_logy requires y>0")
    logy = np.log(y)

    # endpoint slopes in log(y) vs x
    m_lo = (logy[1] - logy[0]) / (x[1] - x[0])
    b_lo = logy[0] - m_lo * x[0]
    m_hi = (logy[-1] - logy[-2]) / (x[-1] - x[-2])
    b_hi = logy[-1] - m_hi * x[-1]

    xq = np.atleast_1d(xq).astype(float)
    out = np.empty_like(xq)

    for i, xv in enumerate(xq):
        if xv < x[0]:
            out[i] = np.exp(m_lo * xv + b_lo)
        elif xv > x[-1]:
            out[i] = np.exp(m_hi * xv + b_hi)
        else:
            out[i] = np.exp(np.interp(xv, x, logy))
    return out

# ----------------------------
# Load Zukic digitized data: use pt["value"] = [lambda_nm, k]
# ----------------------------
with open(ZUKIC_JSON, "r") as f:
    j = json.load(f)

pts = j["datasetColl"][0]["data"]
lambda_nm = np.array([p["value"][0] for p in pts], dtype=float)
k = np.array([p["value"][1] for p in pts], dtype=float)

# sort
idx = np.argsort(lambda_nm)
lambda_nm = lambda_nm[idx]
k = k[idx]

if np.any(k <= 0):
    raise ValueError("k has non-positive entries; check digitization/log-axis settings.")

# ----------------------------
# Plot 1: Input k(λ) with LINEAR interpolation between points
# ----------------------------
lam_dense = np.linspace(lambda_nm.min(), lambda_nm.max(), 2000)
k_dense_linear = interp_linear(lam_dense, lambda_nm, k)

plt.figure()
plt.plot(lambda_nm, k, marker="o", linestyle="None")
plt.plot(lam_dense, k_dense_linear, linestyle="-")
plt.yscale("log")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Extinction coefficient k")
plt.title("Zukic CaF2 k(λ): points + linear interpolation")
plt.grid(True, which="both", alpha=0.3)

# ----------------------------
# Build corrected absorption curve (Zukic shape anchored to Görling bulk α at 193/157 nm)
# ----------------------------
alpha_shape = alpha_from_k(lambda_nm, k)  # cm^-1

# alpha_shape at anchor wavelengths
if ALLOW_EXTRAPOLATION:
    alpha_shape_193 = extrap_logy(193.0, lambda_nm, alpha_shape)[0]
    alpha_shape_157 = extrap_logy(157.0, lambda_nm, alpha_shape)[0]
else:
    alpha_shape_193 = interp_logy(193.0, lambda_nm, alpha_shape)
    alpha_shape_157 = interp_logy(157.0, lambda_nm, alpha_shape)

r193 = alpha_193_bulk / alpha_shape_193
r157 = alpha_157_bulk / alpha_shape_157
log_r193 = np.log(r193)
log_r157 = np.log(r157)

def r_of_lambda(lam):
    lam = np.asarray(lam, float)
    t = (lam - 193.0) / (157.0 - 193.0)   # 0 at 193, 1 at 157
    return np.exp(log_r193 + t * (log_r157 - log_r193))

alpha_corr_curve = alpha_shape * r_of_lambda(lambda_nm)  # cm^-1

# ----------------------------
# Final model evaluated on a dense wavelength grid:
# use logy interpolation (physically better than linear for absorption)
# ----------------------------
if ALLOW_EXTRAPOLATION:
    alpha_dense = extrap_logy(lam_dense, lambda_nm, alpha_corr_curve)
else:
    alpha_dense = interp_logy(lam_dense, lambda_nm, alpha_corr_curve)

L_dense_mm = 10.0 / alpha_dense  # mm

# ----------------------------
# Evaluate final table on your energy grid
# ----------------------------
lambda_grid = energy_to_lambda_nm(energy_eV)

alpha_grid = np.empty_like(lambda_grid)
lam_min, lam_max = float(lambda_nm.min()), float(lambda_nm.max())

for i, lam in enumerate(lambda_grid):
    if lam > lam_max:
        # transparent region beyond Zukic coverage
        alpha_grid[i] = 10.0 / ABSLENGTH_LONG_MM  # since alpha = 10/L_mm
    else:
        if ALLOW_EXTRAPOLATION:
            alpha_grid[i] = extrap_logy(lam, lambda_nm, alpha_corr_curve)[0]
        else:
            if lam < lam_min:
                raise ValueError(f"Need extrapolation at λ={lam:.2f} nm but ALLOW_EXTRAPOLATION=False")
            alpha_grid[i] = interp_logy(lam, lambda_nm, alpha_corr_curve)

L_grid_mm = 10.0 / alpha_grid

# ----------------------------
# Plot 2a: Final ABSLENGTH vs wavelength (dense curve + your grid points)
# ----------------------------
plt.figure()
plt.plot(lam_dense, L_dense_mm, linestyle="-")
plt.plot(lambda_grid, L_grid_mm, marker="o", linestyle="None")
plt.yscale("log")
plt.xlabel("Wavelength (nm)")
plt.ylabel("ABSLENGTH (mm)")
plt.title("Final CaF2 ABSLENGTH vs wavelength (anchored to Görling, shaped by Zukic)")
plt.grid(True, which="both", alpha=0.3)

# Mark anchor wavelengths
plt.axvline(193.0, linestyle="--")
plt.axvline(157.0, linestyle="--")

# ----------------------------
# Plot 2b: Final ABSLENGTH vs energy (this is what your table is indexed by)
# ----------------------------
plt.figure()
plt.plot(energy_eV, L_grid_mm, marker="o", linestyle="-")
plt.yscale("log")
plt.xlabel("Energy (eV)")
plt.ylabel("ABSLENGTH (mm)")
plt.title("Final CaF2 ABSLENGTH table on your energy grid")
plt.grid(True, which="both", alpha=0.3)

plt.show()
```
