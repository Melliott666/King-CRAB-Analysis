```python
import json
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Nexus did not have the absorption length information for Calcium Fluoride and all our optical pathing needs this
# Görling https://www.sciencedirect.com/science/article/pii/S0030401805000490?via%3Dihub has the bulk behavior
alpha_193_bulk = 2.89e-3   # cm^-1
alpha_157_bulk = 17.5e-3   # cm^-1

# Zukic https://opg.optica.org/ao/fulltext.cfm?uri=ao-29-28-4284 has the thin film behavior as a broad spectra
ZUKIC_JSON = Path("../DATA/caf2_zukic_k.json")

#The energy ranges Nexus cares about 
energy_eV = np.array([
    0.900, 1.000, 1.296, 1.683, 2.075,
    2.585, 3.088, 3.709, 4.385, 4.972,
    5.608, 6.066, 6.426, 6.806, 7.135,
    7.401, 7.637, 7.880, 8.217
], dtype=float)

ABSLENGTH_LONG_MM = 1.0e4    # 10 m for wavelengths beyond Zukic max (~230 nm) 
ALLOW_EXTRAPOLATION = True     # allow small extrapolation beyond digitized range
```

```python
#All Helper Functions
def energy_to_lambda_nm(E):
    return 1239.841984 / E

def alpha_from_k(lambda_nm, k):
    # α(cm^-1) = 4π k / λ(cm), λ(cm)=λ_nm*1e-7
    return 4.0 * np.pi * k / (lambda_nm * 1e-7)

def interp_linear(xq, x, y):
    return np.interp(xq, x, y)

def interp_logy(xq, x, y):
    y = np.asarray(y, float)
    return np.exp(np.interp(xq, x, np.log(y)))

def extrap_logy(xq, x, y):
    x = np.asarray(x, float)
    y = np.asarray(y, float)
    logy = np.log(y)

    # endpoint slopes in log(y) vs x
    m_lo = (logy[1] - logy[0]) / (x[1] - x[0])
    b_lo = logy[0] - m_lo * x[0]
    m_hi = (logy[-1] - logy[-2]) / (x[-1] - x[-2])
    b_hi = logy[-1] - m_hi * x[-1]

    xq = np.atleast_1d(xq).astype(float)
    out = np.empty_like(xq)

    for i, xv in enumerate(xq):
        if xv < x[0]:
            out[i] = np.exp(m_lo * xv + b_lo)
        elif xv > x[-1]:
            out[i] = np.exp(m_hi * xv + b_hi)
        else:
            out[i] = np.exp(np.interp(xv, x, logy))
    return out

def get_alpha_at(lam_nm, lambda_nm, alpha_vals):
    if ALLOW_EXTRAPOLATION:
        return extrap_logy(lam_nm, lambda_nm, alpha_vals)[0]
    return interp_logy(lam_nm, lambda_nm, alpha_vals)

def make_dense_lambda(lambda_nm, n=2000):
    return np.linspace(float(np.min(lambda_nm)), float(np.max(lambda_nm)), n)
```

```python
with open(ZUKIC_JSON, "r") as f:
    j = json.load(f)

pts = j["datasetColl"][0]["data"]
lambda_nm = np.array([p["value"][0] for p in pts], dtype=float)
k         = np.array([p["value"][1] for p in pts], dtype=float)

idx = np.argsort(lambda_nm)
lambda_nm = lambda_nm[idx]
k         = k[idx]

if np.any(k <= 0):
    raise ValueError("k has non-positive entries; check digitization/log-axis settings.")

print(f"Loaded {len(lambda_nm)} Zukic points: {lambda_nm.min():.1f}–{lambda_nm.max():.1f} nm")

lam_dense = make_dense_lambda(lambda_nm, n=2000)
k_dense_linear = interp_linear(lam_dense, lambda_nm, k)
```

```python
plt.figure()
plt.plot(lambda_nm, k, marker="o", linestyle="None", label="digitized points")
plt.plot(lam_dense, k_dense_linear, linestyle="-", label="linear interpolation")
plt.yscale("log")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Extinction coefficient k")
plt.title("Zukic CaF2 k(λ)")
plt.grid(True, which="both", alpha=0.3)
plt.legend()
plt.show()
```

```python
# Zukic shape -> alpha_shape (cm^-1)
alpha_shape = alpha_from_k(lambda_nm, k)

# Anchor alpha_shape at 193 and 157 nm
alpha_shape_193 = get_alpha_at(193.0, lambda_nm, alpha_shape)
alpha_shape_157 = get_alpha_at(157.0, lambda_nm, alpha_shape)

r193 = alpha_193_bulk / alpha_shape_193
r157 = alpha_157_bulk / alpha_shape_157
log_r193 = np.log(r193)
log_r157 = np.log(r157)

def r_of_lambda(lam):
    lam = np.asarray(lam, float)
    t = (lam - 193.0) / (157.0 - 193.0)   # 0 at 193, 1 at 157
    return np.exp(log_r193 + t * (log_r157 - log_r193))

alpha_corr_curve = alpha_shape * r_of_lambda(lambda_nm)

# Quick check that we hit the targets
a193 = get_alpha_at(193.0, lambda_nm, alpha_corr_curve)
a157 = get_alpha_at(157.0, lambda_nm, alpha_corr_curve)
print("Anchor check (cm^-1):")
print(f"  alpha_corr(193) = {a193:.4e} (target {alpha_193_bulk:.4e})")
print(f"  alpha_corr(157) = {a157:.4e} (target {alpha_157_bulk:.4e})")
print("Anchor ABSLENGTH (mm):")
print(f"  L(193) = {10.0/alpha_193_bulk:.1f} mm")
print(f"  L(157) = {10.0/alpha_157_bulk:.1f} mm")

# Dense corrected alpha(λ) then L(λ)
if ALLOW_EXTRAPOLATION:
    alpha_dense = extrap_logy(lam_dense, lambda_nm, alpha_corr_curve)
else:
    alpha_dense = interp_logy(lam_dense, lambda_nm, alpha_corr_curve)

L_dense_mm = 10.0 / alpha_dense

# Evaluate table on energy grid
lambda_grid = energy_to_lambda_nm(energy_eV)
alpha_grid = np.empty_like(lambda_grid)

lam_min, lam_max = float(lambda_nm.min()), float(lambda_nm.max())

for i, lam in enumerate(lambda_grid):
    if lam > lam_max:
        alpha_grid[i] = 10.0 / ABSLENGTH_LONG_MM
    else:
        alpha_grid[i] = get_alpha_at(lam, lambda_nm, alpha_corr_curve)

L_grid_mm = 10.0 / alpha_grid
```

```python
plt.figure()
plt.plot(lam_dense, L_dense_mm, linestyle="-", label="corrected model (dense)")
plt.plot(lambda_grid, L_grid_mm, marker="o", linestyle="None", label="Geant4 grid points")
plt.yscale("log")
plt.xlabel("Wavelength (nm)")
plt.ylabel("ABSLENGTH (mm)")
plt.title("Final CaF2 ABSLENGTH vs wavelength (anchored to Görling, shaped by Zukic)")
plt.grid(True, which="both", alpha=0.3)
plt.axvline(193.0, linestyle="--")
plt.axvline(157.0, linestyle="--")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(energy_eV, L_grid_mm, marker="o", linestyle="-")
plt.yscale("log")
plt.xlabel("Energy (eV)")
plt.ylabel("ABSLENGTH (mm)")
plt.title("Final CaF2 ABSLENGTH table on your energy grid")
plt.grid(True, which="both", alpha=0.3)
plt.show()
```

```python
# -------------------------------------------------
# Export ABSLENGTH table to C++ (Geant4 / Nexus)
# -------------------------------------------------

def print_cpp_abslength(L_mm, n_prefix_noabs=2):
    """
    L_mm: array of absorption lengths in mm (for your internal energy grid)
    n_prefix_noabs: how many leading noAbsLength_ entries to print
    """

    print("std::vector<G4double> absLength = {")
    
    # Prefix entries (usually for optPhotMinE_ etc.)
    for i in range(n_prefix_noabs):
        if i < n_prefix_noabs - 1:
            print("  noAbsLength_,")
        else:
            print("  noAbsLength_,")
    
    # Now print actual absorption lengths
    for i, L in enumerate(L_mm):
        if i < len(L_mm) - 1:
            print(f"  {L:.6f}*mm,")
        else:
            print(f"  {L:.6f}*mm")
    
    print("};")


# Call it
print_cpp_abslength(L_grid_mm)
```
